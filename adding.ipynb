```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_excel('/Users/mac/Desktop/—É—á–µ–±–∞/–ø—Ä–æ–µ–∫—Ç/raw_set.xlsx')
df = df.drop(columns=['item_id'])
df
```

```python
df.isnull().sum()
```

```python
df_for = df[['days_active', 'photo_count', 'title_length', 'description_length', 'age', 'page_views', 'contacts', 'favorites']]
df_for
```

```python
corr = df_for.corr()
```

```python
matrix = np.triu(df_for.corr())
plt.figure(figsize=(12, 10))
ax = sns.heatmap(df_for.corr(), annot=True, mask=matrix, cmap='RdBu', center=0, annot_kws={'size': 16})
ax.tick_params(axis='x', which='major', labelsize=15, color='000000')
ax.tick_params(axis='y', which='major', labelsize=15)
```
