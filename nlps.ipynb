```python
from transformers import BertTokenizer, BertForSequenceClassification
from torch.nn.functional import softmax

tokenizer = BertTokenizer.from_pretrained('DeepPavlov/rubert-base-cased')
model = BertForSequenceClassification.from_pretrained('DeepPavlov/rubert-base-cased', num_labels=2)

```

```python
text = "–í–æ–Ω—é—á–∏–π —É–∂–∞—Å–Ω—ã–π –∫–æ—Ç —É–±–µ–π—Ç–µ –µ–≥–æ"

inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)

outputs = model(**inputs)

preds = softmax(outputs.logits, dim=-1)
```

```python
translate = ts.translate_text('–í—Å–µ–º –ø—Ä–∏–≤–µ—Ç', src='ru', dest='en', translator='yandex')
print(translate)
```

```python
def translated(text):
    try:
        translated_text = ts.translate_text(text, src='ru', dest='en', translator='yandex')
        return translated_text
    except Exception as e:
        print(f"Translation failed for: {text}. Error: {e}")
        return text

df['DescriptionEn'] = df['DescriptionRu'].apply(translated)

df
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

df_an = df_check[['DescriptionRu', 'page_views']]
df_an
```

```python
analyzer = SentimentIntensityAnalyzer()

scores = analyzer.polarity_scores('I am a very happy person because I have this cat. She is amazing!')
scores_2 = analyzer.polarity_scores('I hate this cat. She is so ugly and annoying!')
scores_3 = analyzer.polarity_scores('I am selling a cat. Call me: +1234567890')

print(scores)
print(scores_2)
print(scores_3)
```

```python
df_an['positivity'] = df_an['DescriptionEn'].apply(lambda x: analyzer.polarity_scores(x)['compound'])

df_an.head(40)
```
